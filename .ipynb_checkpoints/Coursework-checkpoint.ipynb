```python
import pandas as pd, numpy as np

df = pd.read_csv("Smartphone.csv")
df.columns = df.columns.str.strip().str.replace(" ", "_")

brand_col, model_col, price_col = "brand_name", "model", "price"

skip = {brand_col, model_col}
for c in df.columns:
    if c not in skip and df[c].dtype == "object":
        tmp = pd.to_numeric(df[c].astype(str).str.replace(r"[^\d.]", "", regex=True), errors="coerce")
        if tmp.notna().sum() > 0:
            df[c] = tmp

# fill missing numeric values
num_cols = df.select_dtypes(include="number").columns
df[num_cols] = df[num_cols].fillna(df[num_cols].mean(numeric_only=True))

def recommend_top5():
    b = input("Brand (or any): ").strip().lower()
    d = df.copy()

    if b not in ["", "any"]:
        d = d[d[brand_col].astype(str).str.lower().str.contains(b, na=False)]

    if d.empty:
        return print(" No phones found!")

    # score using numeric columns (exclude price)
    f = d.select_dtypes(include="number").columns.tolist()
    if price_col in f: f.remove(price_col)

    den = (d[f].max() - d[f].min()).replace(0, 1)
    d["Score"] = (((d[f] - d[f].min()) / den).fillna(0)).mean(axis=1)

    show_specs = [c for c in ["ram_capacity","internal_memory","screen_size","battery_capacity","avg_rating"] if c in d.columns][:4]
    show_cols = [brand_col, model_col, price_col, "Score"] + show_specs

    print("\n--- TOP 5 BEST PHONES ---")
    print(d.sort_values("Score", ascending=False)[show_cols].head(5).to_string(index=False))

recommend_top5()
```
